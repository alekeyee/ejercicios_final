```python
# RESOLUCIÓN:
def main(tuple_1):
  tuplas=0
  listas=0
  for identificador in tuple_1:
      if type(identificador) == tuple: # Identifica si es una tupla
        tuplas += 1 #cuenta las tuplas
      if type(identificador) == list: # Identifica si es una lista
        listas +=1 #cuenta las listas
  resultado=tuplas-listas # Resta el total de tuplas con el total de listas
  print(resultado)

main(('k', 'cheers', (8,7), [32.3,-1], (9,)))
```
