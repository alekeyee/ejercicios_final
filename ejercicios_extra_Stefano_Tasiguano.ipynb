```python
def main(pileta):
    piedra = lambda canal: len(canal)
    pileta.sort(key=piedra)
    teatro = dict()
    for ayudante in pileta:
        if len(ayudante) in teatro: teatro[len(ayudante)].append(ayudante)
        else: teatro[len(ayudante)]=[ayudante]
        teatro[len(ayudante)].sort()
    print(teatro)

main(["ab", "abc", "ad"]) 
```
